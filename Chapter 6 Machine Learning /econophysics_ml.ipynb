# Econophysics: Simple ML Example
Predict stock returns from past returns.

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```
np.random.seed(0)
N=1000
price=100+np.cumsum(np.random.normal(0,1,N))
returns=np.diff(np.log(price))
```

```
X,y=[],[]
for i in range(3,len(returns)):
    X.append(returns[i-3:i])
    y.append(returns[i])
X=np.array(X); y=np.array(y)
```

```
Xtr,Xte,ytr,yte=train_test_split(X,y,test_size=0.3)
model=LinearRegression(); model.fit(Xtr,ytr)
yp=model.predict(Xte)
print('MSE=',mean_squared_error(yte,yp))
```

```
plt.plot(yte[:100],label='True'); plt.plot(yp[:100],label='Predicted'); plt.legend(); plt.show()
```
